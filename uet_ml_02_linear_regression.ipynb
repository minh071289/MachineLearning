```python
def visualization(w0, w1, x_train, y_train):
    y_predict = w1 * x_train + w0
    plt.clf()
    plt.figure()
    plt.scatter(x_train, y_train)
    plt.scatter(x_train, y_predict, color='red')
    plt.show()

# training linear regression model
T = 10
learning_rate = 0.001
w0, w1 = torch.rand(2)

def train(x_train, y_train, w0, w1, learning_rate, T):
  for t in range(T):
      y_predict = w1 * x_train + w0
      visualization(w0=w0, w1=w1, x_train=x_train, y_train=y_train)
      # print(f"{y_predict=}")

      # calculate loss
      loss = ((y_predict - y_train) ** 2).mean()

      # dl/dw0 = d(w1 * x + w0 -y)^2 / dw0 = 2(w1 * x + w0 - y)
      # dl/dw1 = d(w1 * x + w0 -y)^2 / dw0 = 2(w1 * x + w0 - y)*x

      # calculate gradient
      w0_grad = 2 * (y_predict - y_train).mean()
      w1_grad = 2 * (x_train * (y_predict - y_train)).mean()

      # update parameter
      w0 = w0 - learning_rate * w0_grad
      w1 = w1 - learning_rate * w1_grad
  return w0, w1
```

```python
# normalization
# x, y have too large values, leading to very large gradients
# we 'normalize' x and y to normalize the gradient

x_train_norm = (x_train - x_train.mean()) / x_train.std()
y_train_norm = (y_train - y_train.mean()) / y_train.std()
```

```python
visualization(w0=w0, w1=w1, x_train=x_train_norm, y_train=y_train_norm)
```

```python
w0, w1 = torch.rand(2)
learning_rate = 0.1
T = 10
train(x_train=x_train_norm,
      y_train=y_train_norm,
      w0=w0, w1=w1,
      learning_rate=learning_rate,
      T=T)
```
